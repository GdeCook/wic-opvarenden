```python
# Definieer taken: (taaknaam, starttijd, duur)
tasks = [
    ('Marktonderzoek', 0.5, 0.5),
    ('Concept Missie en Visie', 1, 0.5),
    ('Financiële Schets', 1.5, 0.5),
    ('Selectie Bestuursleden', 2, 0.5),
    ('Juridisch Advies', 2.5, 0.5),
    ('Ontwerp Deelnemersraad', 3, 0.5),
    ('Stakeholder Communicatie', 3.5, 0.5),
    ('Definitieve Statuten', 4, 0.5),
    ('Bankrekening', 4.5, 0.5),
    ('Website en Communicatie', 5, 0.5),
    ('Fondsenwerving', 5.5, 0.5),
    ('Definitieve Statuten', 6, 0.5),
    ('Inspraakbijeenkomst', 6.5, 0.5),
    ('Eerste Bestuursvergadering', 7, 0.5),
    ('KvK Voorbereiding', 7.5, 0.5),
    ('Formele Registratie', 8, 0.5),
    ('Lancering Website', 8.5, 0.5),
    ('Projectplanning', 9, 0.5),
    ('Ledenwerving', 9.5, 0.5)
]

fig, ax = plt.subplots(figsize=(10, 8))

# Plot taken
for i, (task, start, duration) in enumerate(reversed(tasks)):
    ax.broken_barh([(start, duration)], (i-0.4, 0.8), facecolors='blue')

# Aslabels
ax.set_yticks(np.arange(len(tasks)))
ax.set_yticklabels([task for task, _, _ in reversed(tasks)])

# X-aslimieten en labels
ax.set_xlim(0, 10)
xticks = np.arange(0.5, 10, 1)
ax.set_xticks(xticks)
ax.set_xticklabels(['Sep 1e helft', 'Sep 2e helft', 'Okt 1e helft', 'Okt 2e helft', 'Nov 1e helft', 'Nov 2e helft', 'Dec 1e helft', 'Dec 2e helft', 'Jan 1e helft', 'Jan 2e helft'])

plt.xticks(rotation=45)

# Toon grafiek
plt.title('Tijdpad voor de Oprichting van Stichting "Netwerk Maritieme Bronnen"')
plt.xlabel('Tijd')
plt.show()
```
